# 01 - Extraer frames originales

Objetivo: convertir videos propios de semillas en imagenes `.jpg` clasificadas por carpeta.

```python
from pathlib import Path
import cv2

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
OUTPUT_DIR = PROJECT_ROOT / 'dataset_frames'
OUTPUT_DIR.mkdir(exist_ok=True)

VIDEO_SOURCES = {
    'arbejas': [],
    'arroz': [],
    'frijol': [],
    'maiz_pira': []
}

FRAME_STEP = 15
```

```python
def extract_frames(video_path: Path, label: str, step: int = FRAME_STEP):
    output_label_dir = OUTPUT_DIR / label
    output_label_dir.mkdir(parents=True, exist_ok=True)

    cap = cv2.VideoCapture(str(video_path))
    if not cap.isOpened():
        raise ValueError(f'No se pudo abrir el video: {video_path}')

    saved = 0
    frame_index = 0
    while True:
        ok, frame = cap.read()
        if not ok:
            break
        if frame_index % step == 0:
            output_path = output_label_dir / f'{video_path.stem}_frame_{frame_index:06d}.jpg'
            cv2.imwrite(str(output_path), frame)
            saved += 1
        frame_index += 1
    cap.release()
    return saved

summary = {}
for label, videos in VIDEO_SOURCES.items():
    summary[label] = sum(extract_frames(Path(video), label) for video in videos)

summary
```

# 04 - Evaluar modelo

Objetivo: evaluar `models/modelo_semillas_best.keras` con el split de test.

```python
from pathlib import Path
import numpy as np
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
TEST_DIR = PROJECT_ROOT / 'dataset_augmented' / 'test'
MODEL_PATH = PROJECT_ROOT / 'models' / 'modelo_semillas_best.keras'
CLASSES = ['arbejas', 'arroz', 'frijol', 'maiz_pira']
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
```

```python
model = tf.keras.models.load_model(MODEL_PATH)
test_ds = tf.keras.utils.image_dataset_from_directory(
    TEST_DIR,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_names=CLASSES,
    shuffle=False
)

loss, accuracy = model.evaluate(test_ds)
loss, accuracy
```

```python
y_true = np.concatenate([labels.numpy() for _, labels in test_ds])
probs = model.predict(test_ds)
y_pred = probs.argmax(axis=1)

print(classification_report(y_true, y_pred, target_names=CLASSES))
confusion_matrix(y_true, y_pred)
```

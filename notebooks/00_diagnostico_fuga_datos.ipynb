# 00 - Diagnostico de fuga de datos

Objetivo: definir controles para evitar que imagenes demasiado similares aparezcan simultaneamente en entrenamiento, validacion y prueba.

```python
from pathlib import Path

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
DATASET_FRAMES = PROJECT_ROOT / 'dataset_frames'
DATASET_CLEAN = PROJECT_ROOT / 'dataset_clean'
CLASSES = ['arbejas', 'arroz', 'frijol', 'maiz_pira']

PROJECT_ROOT
```

## Criterios usados

- Separar imagenes por origen/video antes de dividir splits cuando sea posible.
- Balancear clases en `train`, `validation` y `test`.
- Mantener `test` aislado hasta la evaluacion final.
- Revisar manualmente muestras dudosas, borrosas o mal etiquetadas.

```python
def count_images(root: Path):
    rows = []
    for cls in CLASSES:
        folder = root / cls
        count = len(list(folder.glob('*.jpg'))) if folder.exists() else 0
        rows.append((cls, count))
    return rows

count_images(DATASET_FRAMES)
```
